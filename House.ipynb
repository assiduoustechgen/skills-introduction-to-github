```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
df= pd.read_csv('train.csv')
df.drop(['Id','PoolQC','MiscFeature'], axis =1, inplace=True)

#drop PoolQC, MiscFeature and ; dropping Alley and Fence, but we use them as binary possibilities
alleymap = {'Grvl':1, 'Pave':1, np.nan:0}
df['Alley/ No Alley'] = df['Alley'].map(alleymap)
df.drop(['Alley'], axis =1, inplace=True)
df['LotFrontage'].fillna(df['LotFrontage'].mean(), inplace=True)

df['Fence']= df['Fence'].apply(lambda x: 0 if x==0 else 1)

df['MasVnrType']= df['MasVnrType'].apply(lambda x: 1 if x in ['BrkFace','Stone','BrkCmn'] else 0)

df['FireplaceQu']= df['FireplaceQu'].apply(lambda x: 1 if x in ['Gd','TA','Fa','Ex','Po'] else 0)

df[['GarageYrBlt','GarageCond','GarageFinish','GarageQual','GarageType']].value_counts()
df['GarageYrBlt']= df['GarageYrBlt'].fillna(df['GarageYrBlt'].mode()[0])
df['MasVnrArea']= df['MasVnrArea'].fillna(df['MasVnrArea'].mean())
#df['GarageCond'].value_counts().sort_values(ascending=False)
#df['GarageQual'].value_counts().sort_values(ascending=False)

onehot = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
df_num = df.select_dtypes(exclude='object')
df_cat = df.select_dtypes(include= 'object')
encoded_data = onehot.fit_transform(df_cat)
encoded_df = pd.DataFrame(encoded_data, columns=onehot.get_feature_names_out(df_cat.columns))
df_num = df_num.join(encoded_df)

df_num.isnull().sum().sort_values(ascending=False)

rig = Ridge(alpha=0.4)
lr = LinearRegression()
X = df_num.drop('SalePrice', axis=1)
y = df_num['SalePrice']
rig.fit(X,y)
lr.fit(X,y)
y_pred = lr.predict(X)
r2_score = r2_score(y,y_pred)
mse= mean_squared_error(y,y_pred)
print(f"mean squared error is {mse}")
print(rig.score(X,y))
print(f"linear regression score is {r2_score}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score , root_mean_squared_error
df= pd.read_csv('train.csv')
df.drop(['Id','PoolQC','MiscFeature'], axis =1, inplace=True)

#drop PoolQC, MiscFeature and ; dropping Alley and Fence, but we use them as binary possibilities
alleymap = {'Grvl':1, 'Pave':1, np.nan:0}
df['Alley/ No Alley'] = df['Alley'].map(alleymap)
df.drop(['Alley'], axis =1, inplace=True)
df['LotFrontage'].fillna(df['LotFrontage'].mean(), inplace=True)

df['Fence']= df['Fence'].apply(lambda x: 0 if x==0 else 1)

df['MasVnrType']= df['MasVnrType'].apply(lambda x: 1 if x in ['BrkFace','Stone','BrkCmn'] else 0)

df['FireplaceQu']= df['FireplaceQu'].apply(lambda x: 1 if x in ['Gd','TA','Fa','Ex','Po'] else 0)

df[['GarageYrBlt','GarageCond','GarageFinish','GarageQual','GarageType']].value_counts()
df['GarageYrBlt']= df['GarageYrBlt'].fillna(df['GarageYrBlt'].mode()[0])
df['MasVnrArea']= df['MasVnrArea'].fillna(df['MasVnrArea'].mean())
#df['GarageCond'].value_counts().sort_values(ascending=False)
#df['GarageQual'].value_counts().sort_values(ascending=False)

onehot = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
df_num = df.select_dtypes(exclude='object')
df_cat = df.select_dtypes(include= 'object')
encoded_data = onehot.fit_transform(df_cat)
encoded_df = pd.DataFrame(encoded_data, columns=onehot.get_feature_names_out(df_cat.columns))
df_num = df_num.join(encoded_df)

df_num.isnull().sum().sort_values(ascending=False)

rig = Ridge(alpha=0.2)
pl= PolynomialFeatures(degree=1)
lr = LinearRegression()
X = df_num.drop('SalePrice', axis=1)
y = df_num['SalePrice']
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.2,random_state = 1)
x_trainpoly= pl.fit_transform(x_train)
x_testpoly= pl.fit_transform(x_test)
rig.fit(x_trainpoly,y_train)
lr.fit(x_trainpoly,y_train)
y_pred = lr.predict(x_testpoly)
r2_score = r2_score(y_test,y_pred)
mse= mean_squared_error(y_test,y_pred)
rmse = root_mean_squared_error(y_test,y_pred)
print(f"mean squared error is {mse}")
print(f" ridge regression score is {rig.score(x_testpoly,y_test)}")
print(f"linear regression score is {r2_score}")
print(f"root mean squared = {rmse}")
```
